```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
```

## 1_마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 가장 많은 인원을 가진 나이대는?  (0~9: 0, 10~19: 10)

```python
ans = df
ans["age"] = ans["age"] // 10 * 10
ans = ans.groupby("age").size().sort_values(ascending=False)
ans.index[0]
```

```python
# 정답 코드
result = (df.age // 10 * 10).value_counts().index[0]
result
```

## 2_마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 가장 많은 나이대 구간의 인원은 몇명인가?

```python
ans = df.copy()
ans['age'] = ans["age"] // 10 * 10
ans = ans.groupby('age').size().sort_values(ascending=False)
pd.DataFrame(ans)
ans.iloc[0]
```

## 3_나이가 25살 이상 29살 미만인 응답 고객들중 housing컬럼의 값이 yes인 고객의 수는

```python
ans =df
ans = len(ans[((ans["age"] >= 25) & (ans["age"] < 29)) & (ans['housing'] == "yes")])
ans
```

```python
## 정답 코드
# shape 함수: 배열의 크기(각 차원의 길이를 튜플로 표현한 것)
result = df[(df.age >=25) & (df.age<29) & (df.housing =='yes')].shape[0]
print(result)
```

## 4_numeric한 값을 가지지 않은 컬럼들중 unique한 값을 가장 많이 가지는 컬럼은?

```python
ans = df
ans = ans[["job", "marital", "default", "housing", "loan", "contact", "month", "poutcome", "y"]]
ans = ans.nunique().sort_values(ascending=False)
ans = pd.DataFrame(ans)
ans.index[[0, 1]]
```

```python
ans = df
ans = pd.DataFrame(ans)
# exclude 사용해서 제외할 열 선택도 가능(int형과 float형)
# select_columns = ans.select_dtypes(exclude = ['int', 'float']).columns
select_columns = ans.select_dtypes(include = ['object']).columns
new_ans = ans[select_columns]
new_ans = new_ans.nunique().sort_values(ascending=False)

new_ans = pd.DataFrame(new_ans)
new_ans.reset_index(inplace=True)
new_ans.iloc[0:2]['index']
```

```python
# 정답 코드
lst = []
for col in df.select_dtypes(exclude='int'):
  target =  df[col]
  lst.append([col, target.nunique()])

dfs = pd.DataFrame(lst).sort_values(1, ascending=False)
result = dfs[dfs[1] == dfs[1].max()][0].values
print(result)
```

## 5_balance 컬럼값들의 평균값 이상을 가지는 데이터를 ID 값을 기준으로 내림차순 정렬했을 때 상위 100개의 데이터의 balance 값의 평균은?

```python
ans = df[df["balance"] >= df["balance"].mean()].sort_values('ID', ascending=False)
ans.head(100)["balance"].mean()
```

```python
# 정답 코드
result = df[df.balance >= df.balance.mean()].sort_values('ID',ascending=False).head(100).balance.mean()
print(result)
```

## 6_가장 많은 광고를 집행했던 날짜는 언제인가? (데이터 그대로 일(숫자), 달(영문)으로 표기)

## 7_데이터의 job이 unknown 상태인 고객들의 age 컬럼 값의 정규성을 검정하고자 한다. 샤피로 검정의 p-value 값을 구하여라

## 9_y변수와 education 변수는 독립인지 카이제곱검정을 통해 확인하려한다. p-value 값을 출력하라

```python
# 정답 코드
v = pd.crosstab(df.y, df.education)
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(v)  # 카이제곱 통계량, p-value, dof(자유도, degree of freedom), 기대 빈도
display(v)
print(p)
```

## 10_각 job에 따라 divorced/married 인원의 비율을 확인 했을 때 그 값이 가장 높은 값은?

```python
ans = df
ans = ans.groupby(['job', ans['marital']=="divorced"]).size() / ans.groupby(['job', ans['marital']=="married"]).size().sort_values(ascending=False)
ans = pd.DataFrame(ans)
ans.reset_index(inplace = True)
ans = ans[ans["marital"] == True]
ans = ans[0].sort_values(ascending=False)

ans = pd.DataFrame(ans)
ans.iloc[0]
```

```python
# 정답 코드
t = df.groupby(['job', 'marital']).size().reset_index()
pivotdf = t.pivot_table(index='job', columns='marital')[0]
pivotdf = pivotdf.fillna(0)
pivotdf['ratio'] = pivotdf['divorced'] / pivotdf['married']

result = pivotdf.sort_values('ratio').ratio.values[-1]
print(result)
```

```python
# 정답 코드
from sklearn.model_selection import train_test_split

x = train.drop(columns = ['ID', 'y'])
xd = pd.get_dummies(x)
y = train['y']

x_train, x_test, y_train, y_test = train_test_split(xd, y, stratify = y, random_state=1)  # 훈련데이터와 테스트 데이터를 구분할 때, 레이블의 범주 비율에 맞게 하라는 것

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
pred = rf.predict_proba(x_test)

from sklearn.metrics import roc_auc_score, classification_report

print('test roc score: ', roc_auc_score(y_test, pred[:, 1]))

test_pred = rf.predict_proba(pd.get_dummies(test.drop(columns=['ID'])))
submission['predict'] = test_pred[:, 1]

print('submission file')
display(submission.head())
submission.to_csv('submission.csv', index=False)
```
